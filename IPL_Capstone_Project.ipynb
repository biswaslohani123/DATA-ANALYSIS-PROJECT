# Which team won the most matches

```python
df.head()
```

```python
match_wins = df['match_winner'].value_counts()
```

```python
match_wins
```

```python
plt.title('Most Match Win By teams')
plt.xlabel('Match Winner team name')
plt.ylabel('Match Winning count of teams')
sns.barplot(x =  match_wins.index, y = match_wins.values, palette= 'viridis' )
plt.tight_layout()
plt.xticks(rotation=45)
plt.figure(figsize=(12,6))
```

# 10 best bowling figures

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
```

```python
bowling_figure = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
```

```python
bowling_figure
```

```python
bowling_figure.plot(kind='bar')
```
